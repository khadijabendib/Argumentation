```python
Arguments=['a','b','c','e','d','g','h','i','q','s','Q']
list2=[]
list2.append(('a', 'b'))
list2.append(('b', 'c'))
list2.append(('c', 'e'))
list2.append(('e', 'd'))
list2.append(('d', 'g'))
list2.append(('g', 'h'))
list2.append(('Q','h'))
list2.append(('h', 'i'))
list2.append(('i', 'q'))
list2.append(('s', 'q'))
list2
Argument_accepte=[]
Argument_no_accepte=[]
for i in range(len(list2)):
  x=0
  for j in range(len(list2)):
    if i!=j:
      if list2[i][0]==list2[j][1]:
        x=1
  if x==0:
    Argument_accepte.append(list2[i][0])

for argument in Argument_accepte:
    for i in range(len(list2)):
      if list2[i][0]==argument:
        if list2[i][1] not in Argument_no_accepte:
           Argument_no_accepte.append(list2[i][1])
while (len(Argument_accepte)+len(Argument_no_accepte))<len(Arguments):
     for argument in Argument_accepte:
      for i in range(len(list2)):
        if list2[i][0]==argument:
          if list2[i][1] not in Argument_no_accepte and list2[i][1] not in Argument_accepte:
           Argument_no_accepte.append(list2[i][1])
     for arg in Argument_no_accepte:
      for i in range(len(list2)):
        if list2[i][0]==arg:
          if list2[i][1] not in Argument_accepte and list2[i][1] not in Argument_no_accepte:
           Argument_accepte.append(list2[i][1]) 
  # for i in range(len(list2)):
  #    if list2[i][0] not in Argument_accepte and list2[i][0] not in Argument_no_accepte:
       
  #        for arg in Argument_no_accepte:
  #         if list2[i][0]==arg:
  #           if list2[i][1] not in Argument_accepte:
  #             Argument_accepte.append(list2[i][1])
         
```

```python
import networkx as nx
import matplotlib.pyplot as plt

# Créer un graphe dirigé
G = nx.DiGraph()

# Ajouter les noeuds correspondant aux lettres
for lettre in Arguments:
    G.add_node(lettre)

# Ajouter les arêtes entre les noeuds
for i in range(len(list2)):
  G.add_edge(list2[i][0],list2[i][1])


# Dessiner le graphe
pos = nx.circular_layout(G)  # Positionner les noeuds en cercle
node_colors = ['blue' if n in Argument_accepte else 'lightblue' if n in Argument_no_accepte else 'red' for n in G.nodes()]
nx.draw_networkx_nodes(G, pos=pos, node_color=node_colors)
nx.draw_networkx_labels(G, pos)
nx.draw_networkx_edges(G, pos=pos, arrows=True)
plt.show()
```
